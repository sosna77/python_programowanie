```python
class IQDescriptor:
    def __get__(self, instance, owner):
        return instance._IQ
    
    def __set__(self, instance, val):
        if 0 < val < 100:
            instance._IQ = val
        else:
            instance._IQ = 0

class StudentDebil:
    IQ = IQDescriptor()

    def __init__(self, val):
        self.IQ = val

stud = StudentDebil(300)
print(stud.IQ)
stud.IQ = 17
print(stud.IQ)
```
