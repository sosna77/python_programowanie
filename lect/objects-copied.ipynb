````python
import random

class Smartphone:
    def __init__(self, name):
        print('creating smartphone')
        self.name = name
        self.address_book = {}
        self._secret_code = random.randint(1000, 9999)

    def reset(self):
        print(f'reset done to {self.name}')
        self.name = ''
        self.address_book = {}

    def send_data_to_china(self):
        print(f'Sending data to China with the secret code {self._secret_code}')

    def __str__(self):
        return f'SmartPhone(name={self.name}, address_book={self.address_book})'
    
    def __getitem__(self, key):
        return self.address_book[key]
    
    def __setitem__(self, key, value):
        self.address_book[key] = value
    
    def __len__(self):
        return len(self.address_book)


spA = Smartphone("Bob's Phone")
spA.address_book['Alice'] = '123-456-789'
spA.address_book['Charlie'] = '987-654-321'

````

````python
class ElectronicDevice:
    def __init__(self,name='', **kwargs):
        print('ebe')
        self.name=name

    def restart(self):
        print(f'resetin {self.name}')


    __restart = restart


    def update_firmware(self):
        print(f'updating cwyla ciozge for {self.name}')
        self.__restart



class Computer(ElectronicDevice):
    def __init__(self,ram=0, **kwargs):
        super().__init__(*kwargs)
        self.ram=ram
        print(' creationg a komputr')
    def __str__(self):
        return f'Computer name={self.name}, ram={self.ram}'
    def restart(self):
        print(f'saving cwyl ciozge do {self.name}')
        super().restart()



class Camera(ElectronicDevice):
    def __init__(self, width=0, height=0,**kwargs):
        print('creuje kamere')
        super().__init__(**kwargs)
        self.width= width
        self.height = height
    def restart(self):
        print(f'bakuje do kamery zdj do {self.name}')
        super().restart()

class smartphone(Computer, Camera):
    def __init__(self, name, ram, width, height):
        print('creauj smartfona')
        args={'name': name, 'ram': ram, 'width': width, 'height': height}
        super().__init__(**args)
````
