```python
from typing import NamedTuple

class Student(NamedTuple):
    name: str = 'snur'
    age: int = 0
    grade: int  = 17
    email: str = ' twoja.stara'
    classes: list[str] = []
```

```python
stosio = Student(name='Stosio', age=77, grade=1)
```

```python
kiwi = Student(name='Kiwi', age=17, grade=5)
```

```python
stosio.classes.append('hcuj')
stosio.classes.append('smalczyk')
```

```python
kiwi
```

```python
stosio = Student(name='Stosio', age=77, grade=1, classes=[])
stosio.classes.append('hcuj')
stosio.classes.append('smalczyk')
kiwi = Student(name='Kiwi', age=17, grade=5, classes=[])
stosio, kiwi
```

```python
@dataclass
class Student:
    name: str
    age: int = 20
    email: str = ''
    classes: list[str] = field(default_factory=lambda:[])
```

```python
from dataclasses import field
@dataclass
class Student:
    name: str
    age: int = 20
    email: str = ''
    classes: list[str] = field(default_factory=list)
```

```python
alice = Student(name='Alice', age=30)
bob = Student(name='bob', age=7)
alice.classes.append('ten')
alice.classes.append('chuj')
alice.classes.append('stosio')

alice, bob
```

```python
@dataclass

class Student:
    name: str
    age: int = 20
    email: str = ''
    classes: list[str] = field(default_factory=list)
    
    def _post_init__(self):
        if '@' not in self.email:
            raise ValueError('Ty gupia murbo')
```
