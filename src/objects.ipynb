````python
import random

class Smartphone:
    def __init__(self, name):
        print('creating smartphone')
        self.name = name
        self.address_book = {}
        self._secret_code = random.randint(1000, 9999)

    def reset(self):
        print(f'reset done to {self.name}')
        self.name = ''
        self.address_book = {}

    def send_data_to_china(self):
        print(f'Sending data to China with the secret code {self._secret_code}')

    def __str__(self):
        return f'SmartPhone(name={self.name}, address_book={self.address_book})'
    
    def __getitem__(self, key):
        return self.address_book[key]
    
    def __setitem__(self, key, value):
        self.address_book[key] = value
    
    def __len__(self):
        return len(self.address_book)


spA = Smartphone("Bob's Phone")
spA.address_book['Alice'] = '123-456-789'
spA.address_book['Charlie'] = '987-654-321'

````

````python
spA['Alice']
````

````python
spA['Alice'] = '777-777-777'
spA['Alice']
````

````python
len(spA)
````

````python
class ElectronicDevice:
    def __init__(self, name):
        print('Creating electronic device')
        self.name = name

    def restart(self):
        print(f'Restarting {self.name}')

    def update_firmware(self):
        print(f'Updating firmware for {self.name}')
        self.restart()

class Computer(ElectronicDevice):
    def __init__(self, name, ram):
        super().__init__(name)
        print('Creating a computer')
        self.ram = ram

    def restart(self):
        print(f'Saving files on computer {self.name}')
        # żeby wywołać też restart z klasy wyżej
        super().restart()

    def __str__(self):
        return f'Computer name={self.name}, ram={self.ram}'
````

````python
pc7 = Computer('siedem', 7)
pc7.restart()
````

````python
pc7.update_firmware()
````
