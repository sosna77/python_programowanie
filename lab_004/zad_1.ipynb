```python
from time import perf_counter, sleep
from functools import update_wrapper
import random

class TimerDecorator:
    def __init__(self, func=None, *, print_each_call=True):
        print('creating decorator\n')
        self.call_counter = 0
        self.exe_times = []
        self.stats = {}
        self.print_each_call = print_each_call

        if func is not None:
            self.func = func
            update_wrapper(self, self.func)    
        else:
            self.func = None


    def __call__(self, *args, **kwargs):
        if self.func is None:
            self.func = args[0]
            update_wrapper(self, self.func)
            return self
        
        t_start = perf_counter()
        res = self.func(*args, **kwargs)
        t_end = perf_counter()

        t_exe = t_end - t_start

        self.call_counter += 1
        self.exe_times.append(t_exe)
        if self.print_each_call: print(f'Execution time: {t_exe}')
        return res


    def calculate_stats(self):
        print('calculating stats')
        stats = self.stats
        stats['name'] = self.func.__name__
        stats['call_counter'] = self.call_counter
        stats['min_time'] = min(self.exe_times)
        stats['max_time'] = max(self.exe_times)
        stats['mean_time'] = sum(self.exe_times)/self.call_counter

    def print_stats(self):
        print(self.stats,'\n')
        for key, value in self.stats.items():
            print(f'{key}: {value}')


    
@TimerDecorator(print_each_call=True)
def waiting_func():
    t = abs(random.gauss(0.3, 2))
    sleep(t)
waiting_func()
waiting_func()
waiting_func()

waiting_func.calculate_stats()
waiting_func.print_stats()

@TimerDecorator
def fibon(n):
    if n==0: return 1
    if n==1: return 1
    return fibon(n-1) + fibon(n-2)

fibon(3)
fibon(7)
fibon(8)

fibon.calculate_stats()
fibon.print_stats()

```
