```python
def lcg(seed, a, c, m, N):
    x = seed
    for _ in range(N):
        yield x
        x = (a * x + c) % m
    
gen = lcg(seed=1, a=5, c=3, m=16, N=7)

number = next(gen)
print(number)

number = next(gen)
print(number)

number = next(gen)
print(number)

print('\nNow the rest:')
for i in gen:
    print(i)
```

```python
def lcg_zero_one(seed, a, c, m, N):
    x = seed % m
    for _ in range(N):
        yield x/m
        x = (a * x + c) % m
    
gen = lcg_zero_one(seed=1, a=5, c=3, m=16, N=17)

number = next(gen)
print(number)

number = next(gen)
print(number)

number = next(gen)
print(number)

print('\nNow the rest:')
for i in gen:
    print(i)
```

```python
import matplotlib.pyplot as plt
import numpy as np

generator = lcg_zero_one(7, 7, 11, 31, 1000)
arr = np.array([no for no in generator])

xs = arr[::2]
ys = arr[1::2]

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(xs, ys)

plt.show()
```
