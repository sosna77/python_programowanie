```python
class Temperature:
    def __init__(self, temp):
        self._celsius = temp
    
    def __repr__(self):
        return f'Temperature({self._celsius:.1f}°C)' 

    @property
    def celsius(self):
        return self._celsius
    
    @celsius.setter
    def celsius(self, temp):
        if temp < -273.15:
            self._celsius = -273.15
        else:
            self._celsius = temp

    @property
    def fahrenheit(self):
        return self._celsius*1.8 + 32
    
    @fahrenheit.setter
    def fahrenheit(self, temp):
        if temp < -459.67:
            self._celsius = -273.15
        else:
            self._celsius = (temp - 32) / 1.8

    @property
    def kelvin(self):
        return self._celsius + 273.15
    
    @kelvin.setter
    def kelvin(self, temp):
        if temp < 0:
            self._celsius = -273.15
        else:
            self._celsius = temp - 273.15


    # ===== WARIANT A ======

    @property
    def is_freezing(self):
        if self._celsius < 0: return True
        else: return False
    
    # ===== WARIANT B ======

    @property
    def state(self):
        if self._celsius < 0: return 'solid'
        elif self._celsius < 100: return 'liquid'
        else: return 'gas'

    # ===== WARIANT C ======

    @classmethod
    def from_fahrenheit(cls, temp):
        return cls((temp - 32) / 1.8)
    
    @classmethod
    def from_kelvin(cls, temp):
        return cls(temp - 273.15)
    
    # ===== WARIANT D ======

    def __add__(self, other):
        return Temperature(self._celsius + other._celsius)
    
    def __sub__(self, other):
        return Temperature(self._celsius - other._celsius)


t = Temperature(25)
print(repr(t))

print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0
print(t.kelvin)       # 298.15

t.fahrenheit = 32
print(t.celsius)      # 0.0

t.kelvin = 300
print(t.celsius)      # 26.85

t.kelvin = -300      
print(t.kelvin)      # 0

print(t.is_freezing)
print(t.state)


t1 = Temperature.from_fahrenheit(68)
print(repr(t1))
t2 = Temperature.from_kelvin(273.15)
print(repr(t2))

t3 = t1+Temperature.from_kelvin(13)
print(repr(t3))
```
